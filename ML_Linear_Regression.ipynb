```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(1)
X = 2 * np.random.rand(100, 1)
# y = 4 + 3 * X + np.random.rand(100, 1)
y = 4 + 3 * X + 3 * np.random.rand(100, 1)

plt.scatter(X, y)
plt.xlabel("House size (1000 sq. ft)")
plt.ylabel("Price (Lakhs)")
plt.title("House Price Prediction")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X, y)
print("Slop m:", model.coef_)
print("Intercept b:", model.intercept_)

X_test = [[1.0],[1.2],[1.5],[1.8],[2.0],[2.5]]
prediction = model.predict(X_test)
print("Pridicted price (Lakhs): ", prediction[0][0])

y_prid = model.predict(X)

plt.scatter(X, y)
plt.plot(X, y_prid, color="red")
plt.xlabel("House size (1000 sq. ft)")
plt.ylabel("Price (Lakhs)")
plt.title("House Price Prediction")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
mse = mean_squared_error(y, y_prid)
r2_score = r2_score(y, y_prid)

print("Mean Squared Error: ", mse)
print("R2 Score: ", r2_score)
```

```python
residuals = y - y_prid
plt.scatter(X, residuals)
plt.axhline(0, color="blue", linestyle="--")
plt.title("Residual Plot")
plt.show()
```
